```python
from rite_weight import rite_weight 
import numpy as np
```

# rite_weight running guide

## Step 1: Prepare Input Files

## Ensure you have the following NumPy arrays saved as `.npy` files:

## - `trans_from_Rfeat.npy`: Transition from features (shape: `[n_samples, n_features]`)
## - `trans_to_Rfeat.npy`: Transition to features after lag time τ (same shape as above)

## Each row represents a simulation frame (or segment), and each column corresponds to a reduced feature (e.g., tICA).  
## In the current example, a **1D feature** is used for simplicity.

---

### ⏱ Understanding *_Rfeat.npy Construction

### These arrays define the temporal evolution of your system over lag time **τ**:

#### **Lag Time (τ)**:
####   - `trans_from_Rfeat[i,:]` is the feature at time `t`
####   - `trans_to_Rfeat[i,:]` is the feature at time `t + τ`
#### **Note: Row-Wise Alignment**:
####   - Both arrays should be aligned such that each row-pair `(from → to)` corresponds to a valid transition with time difference τ.

## Step 2: Initialize the rite_weight class

## To begin, create a Python dictionary with keys pointing to the file paths of the prepared NumPy arrays:
## - `'trans_from_Rfeat_path'`: Path to `trans_from_Rfeat.npy`  
## - `'trans_to_Rfeat_path'`: Path to `trans_to_Rfeat.npy`
### These files should be the **aligned feature arrays** as described in Step 1.

## Example:

```python
config = {'trans_from_Rfeat_path': './Inputs/trans_from_Rfeat.npy','trans_to_Rfeat_path': './Inputs/trans_to_Rfeat.npy'} #Creating dict
my_model = rite_weight.rite_weight_model(config) # Initialize the rite_weight calss
```

## Step 3: Run rite_weight Iteration

## To run the RiteWeight reweighting process, you need to define the following input parameters:

## Required Parameters:
## - `initial_weights`: A 1D NumPy array representing the initial weights for each segment.  
#### The number of elements must match the number of rows in `trans_from_Rfeat`.

## - `total_iter`:  Number of RiteWeight iterations to run (type: `int`).

## - `clusters`:  Number of clusters (Voronoi centers) used in each iteration.

## - `weight_out_freq`: Frequency (in iterations) at which reweighted segment weights are saved to file.

```python
initial_weights = np.load('./Inputs/weights_in_uniform_microstate_wise.npy')
print(initial_weights.shape)
total_iter = 10
clusters = 10
weight_out_freq = 1
my_model.rite_weight_iter(initial_weights,clusters,total_iter,weight_out_freq)
```

## Output: Reweights.h5: 
## - The estimated segs' weights will be store in a file named `Reweights.h5`. Dataset name: 'weights_out'

### Dataset Dimensions: 

#### - **Number of rows** in `weights_out`: Equal to the number of **segments**
#### - **Number of columns** in `weights_out`: Equal to `total_iterations // weight_out_freq` — since weights are saved **every `weight_out_freq` iterations**.

## Note on Running rite_weight

## - Since `rite_weight` is **iteration-based**, it is recommended to run it as a **background job** using a **job submission system** (e.g., SLURM, PBS) rather than interactively through a Jupyter notebook.

## - This notebook is intended **for demonstration purposes only**.
